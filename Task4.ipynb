```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv("twitter_training.csv")
```

```python
df.columns = ['tweet_id', 'topic', 'sentiment', 'tweet_text']
```

```python
df = df.dropna()
```

```python
sentiment_counts = df['sentiment'].value_counts()
```

```python
plt.figure()
sentiment_counts.plot(kind='bar')
plt.title("Sentiment Distribution on Twitter")
plt.xlabel("Sentiment")
plt.ylabel("Number of Tweets")
plt.show()
```

```python
plt.figure()
sentiment_counts.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel("")
plt.title("Sentiment Percentage")
plt.show()
```

```python
topic_sentiment = df.groupby(['topic', 'sentiment']).size().unstack()

topic_sentiment.plot(kind='bar', stacked=True)
plt.title("Topic-wise Sentiment Distribution")
plt.xlabel("Topic")
plt.ylabel("Number of Tweets")
plt.show()
```

```python
positive_text = " ".join(df[df['sentiment'] == "Positive"]['tweet_text'])

wordcloud = WordCloud(width=800, height=400).generate(positive_text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Positive Tweets WordCloud")
plt.show()
```

```python
df.to_csv("twitter_sentiment_analysis_output.csv", index=False)

print(" Sentiment Analysis & Visualization Completed Successfully")
```
